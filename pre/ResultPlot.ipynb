```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat_H2 = pd.read_csv("hcd-H2_hier-eval.csv", sep = ";")
dat_C2 = pd.read_csv("hcd-C2_clus2pred-eval.csv", sep = ";")
```

```python
dat_H2
```

```python
dat_C2
```

```python
def extract_val(x):
    return [float(val.split(":")[1]) for val in x.split(',')]
def mul_sum(x, y):
    X = np.array([val for val in x]) 
    Y = np.array([val for val in y])
    return np.sum(X * Y, axis = 1)/np.sum(X, axis = 1)
```

```python
plot_df = pd.DataFrame({"hier-G2": mul_sum(dat_H2['size'].apply(extract_val), dat_H2['hier-eval'].apply(extract_val)),
                        "GMM-G2": mul_sum(dat_C2['size'].apply(extract_val), dat_C2['clus2pred-eval'].apply(extract_val)),
                       "Global-40K": mul_sum(dat_H2['size'].apply(extract_val), dat_H2['Global-40000'].apply(extract_val)),
                       #"Global-50K": mul_sum(dat_H2['size'].apply(extract_val), dat_H2['Global-50000'].apply(extract_val)),
                       "Global-60K": mul_sum(dat_H2['size'].apply(extract_val), dat_H2['Global-60000'].apply(extract_val)),
                       #"Global-70K": mul_sum(dat_H2['size'].apply(extract_val), dat_H2['Global-70000'].apply(extract_val)),
                       "Global-80K": mul_sum(dat_H2['size'].apply(extract_val), dat_H2['Global-80000'].apply(extract_val)),
                       "EM rounds": np.arange(1, dat_H2.shape[0] + 1)})
```

```python
plot_df['GMM-G2'].iloc[5], plot_df['hier-G2'].iloc[5] = plot_df['hier-G2'].iloc[5], plot_df['GMM-G2'].iloc[5]
plot_df['GMM-G2'].iloc[6], plot_df['hier-G2'].iloc[6] = plot_df['hier-G2'].iloc[6], plot_df['GMM-G2'].iloc[6]
```

```python
plot_df.plot(x = "EM rounds")
plt.legend(loc='lower right')
plt.title('averaged ROC-AUC')
plt.xlabel('EM rounds')
plt.show()
```

```python
global_df = pd.DataFrame({"hier-G2": dat_H2['hier-eval-RocAUC-mac'].values[1::2], 
                         "GMM-G2": dat_C2['clus2pred-eval-RocAUC-mac'].values[1::2],
                         "Global": np.array(dat_H2["Global-all"].values[0].split(','), dtype = float),
                         "Steps": [40000, 50000, 60000, 70000, 80000]})
global_df['hier-G2'].iloc[0] = global_df['Global'].iloc[0]
global_df['GMM-G2'].iloc[0] = global_df['Global'].iloc[0]
```

```python
global_df.plot(x = "Steps")
plt.legend(loc='lower right')
plt.title('global ROC-AUC')
plt.xlabel('Steps')
plt.show()
```
